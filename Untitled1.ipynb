```python
import pandas as pd # Mengimpor pandas data seperti tabel
import numpy as np
import matplotlib.pyplot as plt # untuk memplot grafik
import seaborn as sns # menyediakan antarmuka untuk membuat grafik statistik
```

```python
df = pd.read_csv('top-rated-coffee-clean.csv') # memuat data dari file
df.head(25) # menampilkan 25 baris teratas
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='roast_level', order = df['roast_level'].value_counts().index, palette='viridis')
plt.title('Jumlah Kopi berdasarkan Tingkat Pemanggangan')
plt.xlabel('Jumlah Kopi')
plt.ylabel('Tingkat Pemanggangan')
plt.show()
```

```python
# prompt: visualisasikan lebih detail

# Membuat figure dan axes dengan ukuran yang lebih besar
fig, ax = plt.subplots(figsize=(12, 8))

# Membuat countplot menggunakan seaborn
# Menggunakan dataFrame df, dengan variabel 'roast_level' pada sumbu y
# Mengurutkan batang berdasarkan jumlah terbanyak (value_counts)
# Menggunakan palet warna 'viridis'
sns.countplot(data=df, y='roast_level', order=df['roast_level'].value_counts().index, palette='viridis', ax=ax)

# Menambahkan judul yang lebih deskriptif
ax.set_title('Distribusi Jumlah Kopi berdasarkan Tingkat Pemanggangan (Roast Level)', fontsize=16)

# Menambahkan label sumbu dengan ukuran font yang lebih besar
ax.set_xlabel('Jumlah Kopi', fontsize=12)
ax.set_ylabel('Tingkat Pemanggangan (Roast Level)', fontsize=12)

# Menambahkan grid horizontal untuk memudahkan pembacaan
ax.xaxis.grid(True, linestyle='--', alpha=0.7)

# Menambahkan teks jumlah di sebelah setiap batang
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=3)

# Mengatur layout agar rapi
plt.tight_layout()

# Menampilkan plot
plt.show()
```

```python
# prompt: buatkan code untuk mencari peminat paling banyak

# Mencari tingkat pemanggangan (roast level) yang paling banyak peminat
roast_counts = df['roast_level'].value_counts()

# Mendapatkan tingkat pemanggangan dengan jumlah terbanyak
most_popular_roast = roast_counts.index[0]
count_most_popular_roast = roast_counts.iloc[0]

print(f"Tingkat pemanggangan yang paling banyak peminat adalah: {most_popular_roast}")
print(f"Jumlah kopi dengan tingkat pemanggangan ini adalah: {count_most_popular_roast}")
```
